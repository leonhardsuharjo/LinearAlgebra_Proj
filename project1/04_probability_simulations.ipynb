```python
#Least square line with alpha and beta
def least_squares_line(x, y):
    x = np.asarray(x); y = np.asarray(y)
    xbar = x.mean(); ybar = y.mean()
    varx = ((x - xbar)**2).mean()
    cov = ((x - xbar)*(y - ybar)).mean()
    alpha = cov / varx
    beta = ybar - alpha * xbar
    return alpha, beta
```
