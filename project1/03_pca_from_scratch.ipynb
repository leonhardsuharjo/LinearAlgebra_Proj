```python
#pca from covariance 
def pca_from_scratch(X, k=2):
    Xs, mu, sigma = standardize(X)
    C = (Xs.T @ Xs) / Xs.shape[0]              # covariance
    eigvals, eigvecs = np.linalg.eigh(C)       # symmetric -> eigh
    idx = np.argsort(eigvals)[::-1]
    eigvals = eigvals[idx]
    eigvecs = eigvecs[:, idx]
    W = eigvecs[:, :k]                         # top-k directions
    Z = Xs @ W                                 # projected data
    explained = eigvals[:k] / eigvals.sum()
    return Z, W, explained
```
